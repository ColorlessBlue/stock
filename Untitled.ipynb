# yfinance

```python
import yfinance as yf

# ティッカーシンボル
nikkei = yf.Ticker("^N225")
dow = yf.Ticker("^DJI")
sp500 = yf.Ticker("^GSPC")
nas = yf.Ticker("^IXIC")
russel = yf.Ticker("^RUT")
sox = yf.Ticker("^SOX")
```

```python
# 株価取得
nikkei_hist = nikkei.history(period="10y")
dow_hist = dow.history(period="10y")
sp500_hist = sp500.history(period="10y")
nas_hist = nas.history(period="10y")
russel_hist = russel.history(period="10y")
sox_hist = sox.history(period="10y")
```

```python
nikkei_hist
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import datetime

plt.figure(figsize=(20, 5))
plt.plot(nikkei_hist.index, stats.zscore(nikkei_hist['Close']), label='Nikkei')
plt.plot(dow_hist.index, stats.zscore(dow_hist['Close']), label='Dow')
plt.plot(sp500_hist.index, stats.zscore(sp500_hist['Close']), label='S&P500')
plt.plot(nas_hist.index, stats.zscore(nas_hist['Close']), label='Nasdaq')
plt.plot(russel_hist.index, stats.zscore(russel_hist['Close']), label='Russel')
plt.plot(sox_hist.index, stats.zscore(sox_hist['Close']), label='Sox')

plt.vlines(x=datetime.datetime.strptime('2022-02-24 00:00:00', '%Y-%m-%d %H:%M:%S'), ymin=stats.zscore(nikkei_hist['Close']).min(), ymax=stats.zscore(nikkei_hist['Close']).max(), color='r', alpha=0.5)
plt.vlines(x=datetime.datetime.strptime('2023-10-07 00:00:00', '%Y-%m-%d %H:%M:%S'), ymin=stats.zscore(nikkei_hist['Close']).min(), ymax=stats.zscore(nikkei_hist['Close']).max(), color='r', alpha=0.5)

plt.xticks(rotation=90)
plt.legend()
plt.show()
```
